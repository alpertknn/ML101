->> To handle a data science project, we will inspect a needed data science lifecycle steps.
### <font color="red">Finding a Solution to Business Problems</font>
->> Should focus on the solution of the problem, should find an algorithm to problem.
<br>
->> The significant point is understanding the problem, the dataset, the developing a solution for that problem. 
<br>
->> The key point is analying the business problem and creating a solution of that problem.
<br>
->> Lets assume that we have a company and our company has some goals like ; Increasing incomes, performance, optimization of processes, and taking strategical decisions. The corporate has some goals and there are some problems are called as business problem.
<br>
->> The problems are solved by approaches. These are used by using algorithms, and tools.
<br>
->> For technical and data based solutions, ofc we can use ml algorithms and tools, the key idea is finding a solution for the given problem, that means we can find a solution by our experiences, analytical skills, not using algorithms and ml tools, too.
<br>
->> Lets assume that we have a sales problem, we should ask ourself like that "what if our sale is this". Thus, we use a predictive model and by asking a question we can think rationally.
### <font color="red"> Random Beginning Question Examples</font>
->> What is the necessity for related business units and corporate ?
<br>
->> By this solution / model or project, which solution will we solve ? (reverse solution), like what is the contribution of ...
<br> 
->> what will be result at the end of the day, so we need to think, what will be the output ?, should calculate project cost, etc.. The important key is that whether the project will contribute the corporate or not.
<br>
->> What is the tangible contrubiton and what is the contribute to business problem ? We should describe the contributions and advantages, etc..
<br>
->> What we want to measure or solve ? =>> On which subject should we measure or solve, etc..
<br>
->> What we want to predict ? =>> Monthly or annual sales, churn of customers on 1 week, etc..
<br>
->> Who will use this model or prediction ? =>> This needs to be cleared at the beginning of project/work. to an example, If we are dealing with sales problem and financial department will use, we can represent as excel form. If end user will use, we need to use mysql or similar structures.
<br>
->> What is the success metric of project ? =>> We should determine the metric of success, as an example %80 can be a success for data scientist but it depends on project or taken decisions.
<br>
->> Who will be responsible to validation or controlling of this process or business problem? 
<br>
->> How long will project take ? 
<br>
->> Which departments will be inside in this project ?

### <font color="red"> Data Preparition</font>
->> After previous steps, we should prepare our dataset to modelization processes.
<br>
->> We should do some data preprocessing steps like dealing with outliers or NA values. We understand data, but we have soem defections on our dataset, so we need to remove this defections from our dataset. 
<br>
->> Variable Standardization, Transormations, etc..
### <font color="red"> Modeling</font>
->> After Data Preparition steps,we can model our problem to predicting car prices.
<br>
->> We should try to learn which variables or conditions are affecting the price ? (Beta values)
<br>
->> We are trying to find coefficients of that model which are affecting the price of the car..
### <font color="red">Evaluation</font>
->> After Modelization process, we should find the success rate of our model.
<br>
->> As we remember, in regression problems, we were using mean squared error metric to find success rate.
<br>
->> In classification methods, we were using accuracy score to find success rate.
<br>
->> The success rates mean is that the measuring of succees rate or comparing  the true and predicted values. 
<br>
->> The end of the day, our main is that reaching the coefficients which are making the difference minimum between true and predicted values.
